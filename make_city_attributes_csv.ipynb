```python
import pandas as pd
```

```python
city_coverage = pd.read_csv('data/city_coverage_geocodes.csv')
city_coverage
```

```python
#They spelled Cincinnati wrong...
city_coverage.loc[11, 'loc'] = 'Cincinnati,OH'
```

```python
#Just make city and state columns
city_coverage['city'] = city_coverage['loc'].apply(lambda s: s.split(',')[0])
city_coverage['state'] = city_coverage['loc'].apply(lambda s: s.split(',')[1])
city_coverage[['city', 'state']]
```

```python
acs_2017 = pd.read_csv('data/acs_2017_5_year_city_level_race_data.csv', encoding = 'latin-1')
acs_2017.head()
```

```python
# http://code.activestate.com/recipes/577305-python-dictionary-of-us-states-and-territories/

states = {
        'AK': 'Alaska',
        'AL': 'Alabama',
        'AR': 'Arkansas',
        'AS': 'American Samoa',
        'AZ': 'Arizona',
        'CA': 'California',
        'CO': 'Colorado',
        'CT': 'Connecticut',
        'DC': 'District of Columbia',
        'DE': 'Delaware',
        'FL': 'Florida',
        'GA': 'Georgia',
        'GU': 'Guam',
        'HI': 'Hawaii',
        'IA': 'Iowa',
        'ID': 'Idaho',
        'IL': 'Illinois',
        'IN': 'Indiana',
        'KS': 'Kansas',
        'KY': 'Kentucky',
        'LA': 'Louisiana',
        'MA': 'Massachusetts',
        'MD': 'Maryland',
        'ME': 'Maine',
        'MI': 'Michigan',
        'MN': 'Minnesota',
        'MO': 'Missouri',
        'MP': 'Northern Mariana Islands',
        'MS': 'Mississippi',
        'MT': 'Montana',
        'NA': 'National',
        'NC': 'North Carolina',
        'ND': 'North Dakota',
        'NE': 'Nebraska',
        'NH': 'New Hampshire',
        'NJ': 'New Jersey',
        'NM': 'New Mexico',
        'NV': 'Nevada',
        'NY': 'New York',
        'OH': 'Ohio',
        'OK': 'Oklahoma',
        'OR': 'Oregon',
        'PA': 'Pennsylvania',
        'PR': 'Puerto Rico',
        'RI': 'Rhode Island',
        'SC': 'South Carolina',
        'SD': 'South Dakota',
        'TN': 'Tennessee',
        'TX': 'Texas',
        'UT': 'Utah',
        'VA': 'Virginia',
        'VI': 'Virgin Islands',
        'VT': 'Vermont',
        'WA': 'Washington',
        'WI': 'Wisconsin',
        'WV': 'West Virginia',
        'WY': 'Wyoming'
}


def find_in_ACS(loc):
    """
    Find the location in the ACS dataset.
    """
    loc = loc.replace('Saint', 'St.')
    loc_split = loc.split(',')
    city = loc_split[0]
    state = loc_split[1]
    state = states[state]
    obs_in_state = acs_2017[acs_2017['Geography'].str.contains(state)]
    obs_in_city =  obs_in_state[obs_in_state['Geography'].str.contains(city)]
    #There are still some obs like West, South. Get the one with shortest length
    return obs_in_city[obs_in_city['Geography'].str.len() == obs_in_city['Geography'].str.len().min()]

#Get the observations corresponding to the cities we have.
rows = []
for loc in list(city_coverage['loc']):
    rows.append(find_in_ACS(loc))
    
relevant_obs = pd.concat(rows).reset_index(drop = True)
relevant_obs.head()
assert len(relevant_obs) == len(city_coverage)
```

```python
#making sure the cities match
for i in range(len(relevant_obs)):
    print(relevant_obs.iloc[i]['Geography'], city_coverage.iloc[i]['loc'])
```

```python
#get population numbers and merge
relevant_obs['white population'] = relevant_obs['Estimate; Not Hispanic or Latino: - White alone'] 
relevant_obs['black population'] = relevant_obs['Estimate; Not Hispanic or Latino: - Black or African American alone']
relevant_obs['asian/pacific islander population'] = relevant_obs['Estimate; Not Hispanic or Latino: - Asian alone'] +\
                                                    relevant_obs['Estimate; Not Hispanic or Latino: - Native Hawaiian and Other Pacific Islander alone']
relevant_obs['hispanic population'] = relevant_obs['Estimate; Hispanic or Latino:']
merged = city_coverage.merge(relevant_obs[['white population', 'black population', 'asian/pacific islander population', 
                                 'hispanic population']], left_index = True, right_index = True)
```
